```python
df_Train["FCatering_Services"] = df_Train["AirlineCompany"].apply(assign_food_rating)

# Combine features for content-based filtering
df_Train["FoodExperience"] = df_Train["AirlineCompany"] + " " + df_Train["Class"]

# Convert text features into TF-IDF matrix
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(df_Train["FoodExperience"])

# Compute cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Select a random user for recommendation
user_index = np.random.choice(df_Train.index)
user_id = df_Train.loc[user_index, "UserID"]

# Get similar airlines based on cosine similarity
similar_scores = list(enumerate(cosine_sim[user_index]))
similar_scores = sorted(similar_scores, key=lambda x: x[1], reverse=True)
top_index = similar_scores[1][0]  # Top similar airline (excluding self)

recommended_airline = df_Train.loc[top_index, "AirlineCompany"]

# Print Recommendation
print(f"Recommendation for user '{user_id}' # based on airline class preference and food experience: >", recommended_airline,"Airlines")
```

**A user who values in-flight food on international flights needs a recommendation.**

```python
df_International = df_Train[df_Train["TypeOfTravel"] == "International"].reset_index(drop=True)

# Define food rating function
def assign_food_rating(airline):
    high_quality = ["Emirates", "Singapore Airlines", "Qatar Airways", "Lufthansa", "Vistara"]
    good_quality = ["Air India", "Go First", "SpiceJet"]
    average_quality = ["IndiGo", "Akasa Air"]
    
    if airline in high_quality:
        return 5  
    elif airline in good_quality:
        return 4  
    elif airline in average_quality:
        return 3  
    else:
        return 2  

# Assign food ratings using .loc[]
df_International.loc[:, "FCatering_Services"] = df_International["AirlineCompany"].apply(assign_food_rating)

# Combine relevant features for recommendation
df_International.loc[:, "InternationalFoodExperience"] = (
    df_International["AirlineCompany"] + " " +
    df_International["Class"] + " " +
    df_International["FCatering_Services"].astype(str)
)

# Convert text features into TF-IDF matrix
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(df_International["InternationalFoodExperience"])

# Compute cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Select a random user for recommendation
user_index = np.random.choice(df_International.index)
user_id = df_International.loc[user_index, "UserID"]

# Get similar airlines based on cosine similarity
similar_scores = list(enumerate(cosine_sim[user_index]))
similar_scores = sorted(similar_scores, key=lambda x: x[1], reverse=True)
top_index = similar_scores[1][0]  # Top similar airline (excluding self)

# Ensure correct indexing by using iloc[]
recommended_airline = df_International.iloc[top_index]["AirlineCompany"]

# Print Recommendation
print(f"Recommendation for user '{user_id}' # for best in-flight food on international flights:", recommended_airline)
```

**A user who prefers Akasa Air due to satisfaction.**

```python
df_Akasa = df_Train[df_Train["AirlineCompany"] == "Akasa Air"]

# Check if Akasa Air exists in the dataset
if df_Akasa.empty:
    print("No data available for Akasa Air.")
else:
    # Extract relevant features for similarity
    df_Train.loc[:, "SatisfactionExperience"] = (
        df_Train["AirlineCompany"] + " " +
        df_Train["Class"] + " " +
        df_Train["Satisfaction"].astype(str)
    )

    # Convert text features into TF-IDF matrix
    vectorizer = TfidfVectorizer(stop_words="english")
    tfidf_matrix = vectorizer.fit_transform(df_Train["SatisfactionExperience"])

    # Compute cosine similarity
    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

    # Get index of an Akasa Air user
    akasa_index = df_Akasa.index[0]

    # Get similar airlines based on cosine similarity
    similar_scores = list(enumerate(cosine_sim[akasa_index]))
    similar_scores = sorted(similar_scores, key=lambda x: x[1], reverse=True)

    # Find the top recommendation excluding Akasa Air itself
    for idx, score in similar_scores[1:]:  # Skip index 0 (itself)
        recommended_airline = df_Train.iloc[idx]["AirlineCompany"]
        if recommended_airline != "Akasa Air":
            user_id = df_Train.iloc[idx]["UserID"]
            break  # Stop at the first valid recommendation

    # Print Recommendation
    print(f"Recommendation for user '{user_id}' # who prefers Akasa Air due to satisfaction,", recommended_airline)
```

**A user who loves IndiGo's time-based efficiency and wants similar airlines.**

```python
df_IndiGo = df_Train[df_Train["AirlineCompany"] == "IndiGo"]

# Check if IndiGo exists in the dataset
if df_IndiGo.empty:
    print("No data available for IndiGo.")
else:
    # Combine relevant features for similarity (time efficiency = punctuality + convenience)
    df_Train.loc[:, "PunctualityExperience"] = (
        df_Train["AirlineCompany"] + " " +
        df_Train["Class"] + " " +
        df_Train["DepartureArrivalTimeConvenient"].astype(str) + " " +
        df_Train["OnlineBoarding"].astype(str)
    )

    # Convert text features into TF-IDF matrix
    vectorizer = TfidfVectorizer(stop_words="english")
    tfidf_matrix = vectorizer.fit_transform(df_Train["PunctualityExperience"])

    # Compute cosine similarity
    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

    # Get index of an IndiGo user
    indigo_index = df_IndiGo.index[0]

    # Get similar airlines based on cosine similarity
    similar_scores = list(enumerate(cosine_sim[indigo_index]))
    similar_scores = sorted(similar_scores, key=lambda x: x[1], reverse=True)

    # Find the top recommendation excluding IndiGo itself
    for idx, score in similar_scores[1:]:  # Skip index 0 (itself)
        recommended_airline = df_Train.iloc[idx]["AirlineCompany"]
        if recommended_airline != "IndiGo":
            user_id = df_Train.iloc[idx]["UserID"]
            break  # Stop at the first valid recommendation

    # Print Recommendation
    print(f"Recommendation for user '{user_id}' # who loves IndiGo's time-based efficiency,", recommended_airline)
```

**A user who wants the top 5 international airlines based on satisfaction.**

```python
satisfaction_mapping = {"Dissatisfied": 1, "Neutral": 2, "Satisfied": 3}

# Convert Satisfaction to numeric values
df_Train["Satisfaction"] = df_Train["Satisfaction"].map(satisfaction_mapping)

# Filter only international flights
df_International = df_Train[df_Train["TypeOfTravel"] == "International"]

# Find the top 5 international airlines based on satisfaction scores
top_5_airlines = (
    df_International.groupby("AirlineCompany")["Satisfaction"]
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

# Print Recommendation
print("Recommendation for a user who wants the top 5 international airlines based on satisfaction:")
print(top_5_airlines)
```
