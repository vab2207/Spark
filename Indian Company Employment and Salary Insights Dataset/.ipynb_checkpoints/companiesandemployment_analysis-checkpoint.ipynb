```python

from pyspark.sql.functions import regexp_replace, when, col, length

# Step 1: Clean salary strings and handle suffixes like k, L, Cr
data = (
    data.withColumn("Avg_salary_clean", regexp_replace(col("Avg_salary"), "₹|,|\\s", ""))  # remove ₹, commas, spaces
    .withColumn(
        "Avg_salary_clean",
        when(col("Avg_salary_clean").endswith("k"), regexp_replace(col("Avg_salary_clean"), "k", "").cast("double") * 1000)
        .when(col("Avg_salary_clean").endswith("L"), regexp_replace(col("Avg_salary_clean"), "L", "").cast("double") * 100000)
        .when(col("Avg_salary_clean").endswith("Cr"), regexp_replace(col("Avg_salary_clean"), "Cr", "").cast("double") * 10000000)
        .otherwise(col("Avg_salary_clean").cast("double"))
    )
)

# Step 2: Verify results
data.select("Company_name", "Avg_salary", "Avg_salary_clean").show(20)
```

```python
#Top 10 Highest rated companies

pdf = data.select("Company_name", "Ratings").toPandas()

# Drop missing or invalid values
pdf = pdf.dropna(subset=["Ratings"])
pdf = pdf[pdf["Ratings"] != "--"]

# Convert ratings to numeric safely
pdf["Ratings"] = pd.to_numeric(pdf["Ratings"], errors="coerce")

# Sort and select top 10
top_rated = pdf.sort_values(by="Ratings", ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x="Ratings", y="Company_name", data=top_rated)
plt.title("Top 10 Highest Rated Companies", fontsize=14)
plt.xlabel("Average Rating")
plt.ylabel("Company Name")
plt.show()
```

```python
# Companies having > 2.5 ratings (Below avg)

pdf = data.select("Company_name", "Ratings").toPandas()

# Drop null or invalid ratings
pdf = pdf.dropna(subset=["Ratings"])
pdf = pdf[pdf["Ratings"] != "--"]

# Convert ratings to numeric safely
pdf["Ratings"] = pd.to_numeric(pdf["Ratings"], errors="coerce")

# Filter companies with low ratings
low_rated = pdf[pdf["Ratings"] < 2.5].sort_values(by="Ratings", ascending=True)

# Keep only first 15 for readability
low_rated = low_rated.head(15)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="Ratings", y="Company_name", data=low_rated)
plt.title("Companies with Ratings Below 2.5", fontsize=14)
plt.xlabel("Average Rating")
plt.ylabel("Company Name")
plt.xlim(0, 5)
plt.show()
```

```python
import re

# Define to_num function for salary cleaning
def to_num(x):
    if isinstance(x, str):
        x = re.sub(r'[^\d.]', '', x)
        return float(x) if x else None
    return x

# Convert salary column into numeric form
df['Avg_salary_num'] = df['Avg_salary'].apply(to_num)

# Group by rating category (rounded to 1 decimal place)
df['Ratings_group'] = df['Ratings'].round(1)
avg_salary_group = df.groupby('Ratings_group')['Avg_salary_num'].mean().dropna()

plt.figure(figsize=(12,6))
avg_salary_group.plot(kind='bar')
plt.title('Average Salary vs Ratings Group')
plt.xlabel('Ratings Group')
plt.ylabel('Average Salary (INR)')
plt.tight_layout()
plt.show()
```

```python
# Normalize values to bring them to the same scale (0 to 1)
df['norm_rating'] = (df['Ratings'] - df['Ratings'].min()) / (df['Ratings'].max() - df['Ratings'].min())
df['norm_salary'] = (df['Avg_salary_num'] - df['Avg_salary_num'].min()) / (df['Avg_salary_num'].max() - df['Avg_salary_num'].min())

# Handle missing values safely
df['norm_rating'] = df['norm_rating'].fillna(0)
df['norm_salary'] = df['norm_salary'].fillna(0)

# Combined “badness score” → Lower rating + lower salary
# You can add weights: rating has 60% importance, salary has 40%
df['badness_score'] = (1 - df['norm_rating'])*0.6 + (1 - df['norm_salary'])*0.4

# Get worst 5 companies
worst_5 = df[['Company_name', 'Ratings', 'Avg_salary_num', 'badness_score']].sort_values(by='badness_score', ascending=False).head(5)

print("Worst 5 Companies (Based on Ratings & Salaries):")
display(worst_5)
```

```python
# Identify top 10 companies by total job availability
if 'df' in locals() and 'Total_jobs_available' in df.columns:
    def parse_jobs(x):
        try:
            x = str(x).lower().replace('k','000').replace('m','000000').replace(',','')
            return float(x)
        except:
            return None
    df['Total_jobs_num'] = df['Total_jobs_available'].apply(parse_jobs)
    top_jobs = df.sort_values('Total_jobs_num', ascending=False).head(10)[['Company_name','Total_jobs_num','Ratings']]
    print("Top 10 Companies with Most Job Openings:")
    display(top_jobs)
```

```python
#Companies need to grow in terms of total benefits


df['Total_benefits_num'] = pd.to_numeric(df['Total_benefits'], errors='coerce')

# Sort by total benefits (ascending)
lowest_benefit_companies = (
    df[['Company_name', 'Total_benefits_num']]
    .dropna()
    .sort_values(by='Total_benefits_num', ascending=True)
    .head(5)
)

print("🏢 Companies That Need to Grow in Terms of Total Benefits:")
display(lowest_benefit_companies)
```
